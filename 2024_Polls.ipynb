# Stage 2 – Data Cleaning and Pre-processing

```python
# -------------------------------
# 1. Handle Missing Values
# -------------------------------
for col in num_cols:
    if abs(df[col].skew()) < 1:   # roughly symmetric
        df[col] = df[col].fillna(df[col].mean())
    else:                         # skewed
        df[col] = df[col].fillna(df[col].median())

# Categorical → mode imputation + convert to category
cat_cols = ['pollster_name','sponsor_names','sponsor_classifications',
            'partisanship','internal','state','media?','university?']

for col in cat_cols:
   df[col] = df[col].fillna(df[col].mode()[0]).astype('category')
```

```python
# -------------------------------
# 2. Remove Duplicates
# -------------------------------
print("Duplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# -------------------------------
# 3. Outlier Treatment
# -------------------------------
# Exclude unique IDs and boolean columns as outlier treatment is not meaningful for them
# First, let's get all numerical columns (including booleans by default in pandas dtypes)
all_numeric_cols = df.select_dtypes(include=np.number).columns

# Now, filter out boolean columns and the identifier 'pollster_rating_id'
num_cols_for_outlier_treatment = [
    col for col in all_numeric_cols
    if col != 'pollster_rating_id' and not pd.api.types.is_bool_dtype(df[col])
]

for col in num_cols_for_outlier_treatment:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))
```

```python
# -------------------------------
# 4. Skewness Check & Transform
# -------------------------------
exclude_cols = ['pollster_rating_id','tracking','has_prez?','has_generic?',
                'has_senate?','has_house?','media_or_university']
numeric_cols = [col for col in num_cols if col not in exclude_cols]

print("Skewness:\n", df[numeric_cols].skew())

for col in numeric_cols:
    if abs(df[col].skew()) > 1:
        df[col + '_log'] = np.log1p(df[col].astype('float64'))
```

```python
# -------------------------------
# 5. Convert Data Types
# -------------------------------
df['start_date'] = pd.to_datetime(df['start_date'], format='%m/%d/%y', errors='coerce')
df['end_date']   = pd.to_datetime(df['end_date'], format='%m/%d/%y', errors='coerce')
```

```python
# -------------------------------
# 6. Feature Engineering
# -------------------------------
df['Year'] = df['start_date'].dt.year
df['Month'] = df['start_date'].dt.month
df['PollLength'] = (df['end_date'] - df['start_date']).dt.days

# Final check
print(df.info())
print(df.head())
```

```python
# 3.11 Line Chart: Poll Counts Over Time
polls_per_month = df.groupby(['Year','Month']).size().reset_index(name='PollCount')
polls_per_month['Date'] = pd.to_datetime(polls_per_month[['Year','Month']].assign(DAY=1))

plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='PollCount', data=polls_per_month, marker='o')
plt.title("Poll Counts Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Polls")
plt.show()
```
